What sort of distribution do the second and forth features display? With that in mind, explain the outliers, or lack thereof, in terms of what you observe from the histograms

### Analysis:
Both Feature 1(the second) and Feature 3(the forth) have a lot of outliers.
<br>
Both Feature 1(the second) and Feature 3(the forth) have a distribution of "skewed left".
<br>
$$ lowerInnerFence = 1stQuartile - 1.5 \times (3rdQuartile - 1stQuartile)$$
Anything fallen beneath the lower_inner_fence becomes an outlier. In a sharply skewed left distribution, there are a large number of occurrences in the upper value cells (right side) and very few in the lower value cells (left side). So, the data on the very left becomes outliers in the box plot.
